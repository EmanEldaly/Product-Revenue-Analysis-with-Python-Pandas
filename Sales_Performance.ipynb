```python
import pandas as pd
import numpy as np
```

```python
sal=pd.read_csv('/sales_data.csv')
```

```python
sal.head() #Display the first 5 records
```

```python
sal.describe() #Descriptive statistics
```

```python
sal.info() #Data Info
```

```python
for unique_category in sal['Category'].unique(): #List of unique categories
    print(unique_category)
```

```python
avg=sal['Total Revenue'].mean() #Average total revenue
print(f'Average revenue: {avg}')
```

```python
total=sal['Total Revenue'].mean() # Sum of total revenue
print(f'Total revenue: {total}')
```

```python
maxrev=sal['Total Revenue'].max() #Product with highest revenue
sal[sal['Total Revenue']==maxrev][['Product Name','Total Revenue']]
```

```python
maxquant=sal['Quantity Sold'].max() #Product with highest sales
sal[sal['Quantity Sold']==maxquant][['Product Name', 'Quantity Sold']]
```

```python
catsal=sal.groupby('Category') #Average revenue by category
catsal['Total Revenue'].mean()
```

```python
product_count=sal['Category'].value_counts().reset_index() #number of prudycts in each category
product_count.columns = ['Category', 'Product Count']
print(product_count)
```

```python
sal['Performance'] = np.where(sal['Total Revenue'] > 50000, 'High', 'Low') #High and low performance
print (sal[['Product Name', 'Performance']])
```

```python
total_products = len(sal)
print(f"Total number of products:", total_products)
```

```python
percount=sal['Performance'].value_counts().reset_index()
percount.columns =['Performance','Count']
print (percount)
```
